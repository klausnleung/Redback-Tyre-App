```python
"""
7 tyres
6 runs
2 rim widths per tyre
- 3 per totaling 6 (trans, etc)
"""
```

```python
import os
import pandas as pd
```

```python
meta = []

for f in os.listdir('data'):
    if f.endswith('.txt'):
        with open(f'data/{f}', 'r') as dat:
            meta.append(dat.read())
meta
```

```python
df = pd.DataFrame(columns=['Run', 'Time', 'Tyre_Name', 'Tyre_Specs', 'Rim_Width', 'ID'])

for i, run in enumerate(meta):
    run_i = run.find('Run')
    time_i = run.find('Time')
    tyre_i = run.find('Tire_Name')
    spec_i = run[tyre_i:].find(',')  # from end
    rim_i = run.find('Rim_Width')
    id_i = run.find('ID')
    df  = df.append({'Run': run[run_i+4:run_i+7], 
                     'Time': run[time_i+5:time_i+10],
                     'Tyre_Name': run[tyre_i+10:tyre_i+run[tyre_i:].find(' ')],
                     'Tyre_Specs': run[spec_i-11:spec_i],
                     'Rim_Width': run[rim_i+10:rim_i+13],
                     'ID': run[id_i+3:id_i+10]},
                    ignore_index=True)
    # df  = df.append({'Run': i, 
    #                  'Time': i,
    #                  'Tyre_Name': i,
    #                  'Tyre_Specs': i,
    #                  'Rim_Width': i,
    #                  'ID': i},
    #                 ignore_index=True)
```
